# TP1 : Initiation au traitement d'images

```python
import cv2 as cv
import matplotlib.pyplot as plt
import os
import pandas as pd
```

### 1. Lire une image A en couleur de format JPEG

```python
A = cv.imread('./picnic.jpeg')
plt.imshow(cv.cvtColor(A, cv.COLOR_BGR2RGB))
plt.show()
```

### 2. L'afficher sur l'écran

```python
A_resized = cv.resize(A, (0,0), fx=0.5, fy=0.5)
cv.imshow('ducks', A_resized)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 3. Afficher ses dimensions

```python
dimensions = A.shape
height, width, channels = dimensions
print("Les dimensions de l'image sont : ")
print(f" - Hauteur : {height} pixels.")
print(f" - Largeur : {width} pixels.")
print(f" - Nombre de canaux : {channels}.")
```

### 4. Sauvegarder l'image sous format PNG et TIFF

```python
done1 = cv.imwrite('./picnic_png.png', A)
done2 = cv.imwrite('./picnic_tif.tif', A)
if (done1 and done2):
    print("Images sauvegardées.")
```

### 5. Afficher leurs dimensions et leurs tailles

```python
A_png = cv.imread('./picnic_png.png')
A_tif = cv.imread('./picnic_tif.tif')

dimensions_png = A_png.shape
dimensions_tif = A_tif.shape

size_png = os.path.getsize('./picnic_png.png')
size_tif = os.path.getsize('./picnic_tif.tif')
size_jpeg = os.path.getsize('./picnic.jpeg')

form = ['PNG', 'JPEG', 'TIFF']
dim = [dimensions_png, dimensions, dimensions_tif]
size = [size_png, size_jpeg, size_tif]
for i in range(3):
    if size[i] >= 1024*1024:
            size[i] = f"{size[i]/1024/1024:.2f} MB"
    else:
            size[i] = f"{size[i]/1024:.2f} KB"
data = {'Format': form,
        'Dimensions': dim,
        'Taille': size}
        
df = pd.DataFrame(data)
display(df.style.hide(axis="index"))
```

### 6. Comparer les 3 formats

   On remarque que : 
- Taille de l'image TIFF > Taille de l'image PNG > Taille de l'image JPEG, car :

    - Le format **JPEG** sacrifie la qualité de l'image pour la compression.
    - Le format **TIFF** est un format sans perte qui préserve la qualité de l'image.
    - Le format **PNG** prend en charge la compression sans perte.


- Les 3 fichiers ont les memes dimensions.

### 7. Afficher A en niveaux de rouge, niveaux du vert et en niveaux du bleu

```python
b, g, r = cv.split(A)

fig, ax = plt.subplots(2, 3, figsize=(12, 8))

ax[0, 0].imshow(b, cmap='Blues')
ax[0, 0].set_title('A en niveaux de bleu')

ax[0, 1].imshow(g, cmap='Greens')
ax[0, 1].set_title('A en niveaux de vert')

ax[0, 2].imshow(r, cmap='Reds')
ax[0, 2].set_title('A en niveaux de rouge')

ax[1, 0].imshow(b, cmap='gray')
ax[1, 0].set_title('A en niveaux de bleu (en gris)')

ax[1, 1].imshow(g, cmap='gray')
ax[1, 1].set_title('A en niveaux de vert (en gris)')

ax[1, 2].imshow(r, cmap='gray')
ax[1, 2].set_title('A en niveaux de rouge (en gris)')

plt.show()
```

### 8. Reconstruire l'image initiale à partir de ces 3 niveaux.

```python
A_reconstruite = cv.merge((b, g, r))
plt.imshow(cv.cvtColor(A_reconstruite, cv.COLOR_BGR2RGB))
plt.show()
```

### 9. Convertir A en niveaux de gris.

```python
A_gris = cv.cvtColor(A, cv.COLOR_BGR2GRAY)
plt.imshow(A_gris, cmap='gray')
plt.show()
```

### 10. Convertir A en binaire

```python
_, A_binaire = cv.threshold(A_gris, 127, 255, cv.THRESH_BINARY)
# Display the binary image
plt.imshow(A_binaire, cmap='gray')
plt.show()
```

### 11 Trouver l'image inverse de A. L'afficher

```python
A_inv = 255 - A

plt.imshow(A_inv)
plt.show()
```
