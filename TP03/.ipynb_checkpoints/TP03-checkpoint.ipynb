## TP03 : Transformée de Fourier et Filtrage Fréquentiel

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
cameraman = cv.imread('cameraman.jpg',0)
trui = cv.imread('trui.png',0)
```

#### 1. Trouver la TF de l’image cameraman . Afficher l’image module et celle de l’argument.

```python
# Appliquer la TF
f1 = np.fft.fft2(cameraman)
# Faire le décalage
fshift1 = np.fft.fftshift(f1)
magnitude_spectrum1 = 20*np.log(np.abs(fshift1))
phase_spectrum1 = np.angle(fshift1)

fig = plt.figure(figsize=(18, 5))
gs = fig.add_gridspec(1, 3)

ax1 = fig.add_subplot(gs[0, 0:2])
im1 = ax1.imshow(magnitude_spectrum1, cmap='gray')
ax1.set_title("Spectre de module")
ax1.set_xticks([]), ax1.set_yticks([])
plt.colorbar(im1, ax=ax1)

ax2 = fig.add_subplot(gs[0, 2])
im2 = ax2.imshow(phase_spectrum1, cmap='gray')
ax2.set_title('Spectre de phase')
ax2.set_xticks([]), ax2.set_yticks([])
plt.colorbar(im2, ax=ax2)

plt.subplots_adjust(wspace=0.05)
plt.show()
```

#### 2. Trouver la TF de l’image trui . Afficher l’image module de et celle de l’argument.

```python
# Appliquer la TF
f2 = np.fft.fft2(trui)
# Faire le décalage
fshift2 = np.fft.fftshift(f2)
magnitude_spectrum2 = 20*np.log(np.abs(fshift2))
phase_spectrum2 = np.angle(fshift2)

fig = plt.figure(figsize=(18, 5))
gs = fig.add_gridspec(1, 3)

ax1 = fig.add_subplot(gs[0, 0:2])
im1 = ax1.imshow(magnitude_spectrum2, cmap='gray')
ax1.set_title("Spectre de module")
ax1.set_xticks([]), ax1.set_yticks([])
plt.colorbar(im1, ax=ax1)

ax2 = fig.add_subplot(gs[0, 2])
im2 = ax2.imshow(phase_spectrum2, cmap='gray')
ax2.set_title('Spectre de phase')
ax2.set_xticks([]), ax2.set_yticks([])
plt.colorbar(im2, ax=ax2)

plt.subplots_adjust(wspace=0.05)
plt.show()
```

#### 3. Construire une 3ème image dont le module de sa TF est égal à celui de cameraman et son argument est celui de l'image trui. Conclusion.

```python
# Reconstruction d'une nouvelle image avec le module de cameraman et la phase de trui
magnitude_new = np.abs(fshift1)
phase_new = np.angle(fshift2)
new_spectrum = magnitude_new * np.exp(1j * phase_new)
f_new = np.fft.ifftshift(new_spectrum)
new_img = np.fft.ifft2(f_new)
new_img = np.abs(new_img)

# Affichage de la nouvelle image
plt.imshow(new_img, cmap='gray')
plt.title('Nouvelle image avec le module de cameraman et la phase de trui'), plt.xticks([]), plt.yticks([])
plt.show()
```

#### 4. Générer la TF d’un filtre passe-bas de même taille que l'image cameraman. Appliquer ce filtre à l’image dans le domaine fréquentiel. Afficher le résultat. Faites varier la fréquence de coupure $(U_0,V_0)$. Conclusions.

```python
rows, cols = cameraman.shape
crow, ccol = rows//2, cols//2

for d in [10,20, 30,40, 60]:
    mask = np.zeros((rows, cols), np.uint8)
    mask[crow-d:crow+d, ccol-d:ccol+d] = 1
    masked = fshift1 * mask
    f_ishift = np.fft.ifftshift(masked)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.abs(img_back)

    # Affichage 
    plt.subplot(121)
    plt.imshow(mask, cmap='gray')
    plt.title(f'Filtre passe-bas (U0,V0)=({d},{d})')
    plt.xticks([]), plt.yticks([])

    plt.subplot(122)
    plt.imshow(img_back, cmap='gray')
    plt.title(f'Image filtrée (U0,V0)=({d},{d})')
    plt.xticks([]), plt.yticks([])

    plt.tight_layout()
    plt.show()
```

**Conclusion:**

   Dans le contexte du filtrage dans le domaine fréquentiel, la fréquence de coupure représente la fréquence maximale qui est autorisée à traverser un filtre. Un filtre passe-bas permet aux composantes de basses fréquences de passer à travers, tout en bloquant les composantes de hautes fréquences. Par conséquent, en augmentant la fréquence de coupure d'un filtre passe-bas, plus de composantes de hautes fréquences peuvent passer, ce qui conduit à une image plus nette avec plus de détails. À l'inverse, en diminuant la fréquence de coupure d'un filtre passe-bas, plus de composantes de hautes fréquences sont bloquées, ce qui conduit à une image plus lisse avec moins de détails.

#### 5. Refaire le même travail demandé avec un filtre passe-haut.

```python
rows, cols = cameraman.shape
crow, ccol = rows//2, cols//2
for d in [5,10,20, 30,40, 60]:
    mask = np.ones((rows, cols), np.uint8)
    mask[crow-d:crow+d, ccol-d:ccol+d] = 0
    masked = fshift1 * mask
    f_ishift = np.fft.ifftshift(masked)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.abs(img_back)

    plt.subplot(121)
    plt.imshow(mask, cmap='gray')
    plt.title(f'Filtre passe-haut (U0,V0)=({d},{d})')
    plt.xticks([]), plt.yticks([])

    plt.subplot(122)
    plt.imshow(img_back, cmap='gray')
    plt.title(f'Image filtrée (U0,V0)=({d},{d})')
    plt.xticks([]), plt.yticks([])

    plt.tight_layout()
    plt.show()
```

#### 8. Translater cet histogramme à droite et à gauche, afficher à chaque fois l’image obtenue.

```python
# Compute the histogram of the image
hist, bins = np.histogram(cameraman.flatten(), 256, [0, 256])

# Compute the cumulative distribution function of the histogram
cdf = hist.cumsum()

# Compute the normalization factor
norm_factor = (cameraman.shape[0] * cameraman.shape[1]) - 1

# Compute the new minimum and maximum intensity values
new_min = 0
new_max = 255

# Compute the new intensity values using linear interpolation
new_intensities = np.interp(np.arange(256), np.around((cdf / norm_factor) * 255), np.arange(256)) 

# Compute the new histogram
new_hist = np.zeros(256)
for i in range(256):
    new_hist[i] = np.sum(hist[np.where(new_intensities == i)])

# Shift the histogram to the left or right
shift = 50
new_hist_left = np.roll(new_hist.copy(), shift)
new_hist_right = np.roll(new_hist.copy(), -shift)

# Compute the new image using the shifted histograms
new_image_left = np.zeros_like(cameraman)
for i in range(256):
    new_image_left[cameraman == i] = new_intensities[i] + shift
new_image_left[new_image_left > new_max] = new_max
new_image_left[new_image_left < new_min] = new_min

shifted_bins = np.arange(256) - shift
shifted_bins[shifted_bins < 0] = 0

new_hist_right = np.zeros_like(new_hist)
for i in range(256):
    new_hist_right[shifted_bins[i]] += new_hist[i]

# Compute the new image using the shifted histograms
new_image_right = np.zeros_like(cameraman)
for i in range(256):
    new_image_right[cameraman == i] = new_intensities[shifted_bins[i]]

new_image_right[new_image_right > new_max] = new_max
new_image_right[new_image_right < new_min] = new_min


# Display the original and the shifted images
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(cameraman, cmap='gray')
ax[0].set_title('Image originale')
ax[1].imshow(new_image_left, cmap='gray')
ax[1].set_title('Image apres la translation à gauche')
ax[2].imshow(new_image_right, cmap='gray')
ax[2].set_title('Image apres la translation à droite')

plt.show()
```

#### 9. Appliquer l’égalisation de son histogramme. Afficher l’image traitée ainsi que son nouvel histogramme

```python
img_thumb_pout = cv.imread('thumb_pout.png', cv.IMREAD_GRAYSCALE)
eq_img = cv.equalizeHist(img_thumb_pout)

hist, bins = np.histogram(img_thumb_pout.flatten(), 256, [0, 256])
eq_hist, eq_bins = np.histogram(eq_img.flatten(), 256, [0, 256])

plt.subplot(1, 2, 1)
plt.imshow(img_thumb_pout, cmap='gray')
plt.title('Image originale')

plt.subplot(1, 2, 2)
plt.imshow(eq_img, cmap='gray')
plt.title('Image égalisée')

plt.show()

plt.subplot(1, 2, 1)
plt.bar(bins[:-1], hist, width=1)
plt.title('Histogram original')

plt.subplot(1, 2, 2)
plt.bar(eq_bins[:-1], eq_hist, width=1)
plt.title('Histogram égalisé')

plt.show()
```
